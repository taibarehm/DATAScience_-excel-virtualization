```python
"""Do an analysis of â€˜sexâ€™ and â€˜smokerâ€™ and comment on whether females smoke more 
than men or men smoke more than females, in max 30 words
with logical method
"""
list3=zip(excel_red.sex,excel_red.smoker)
print(list3)
femaelwhoSmoke=0
femaelwhoDontsmoke=0
maelwhoSmoke=0
maelwhoDontsmoke=0
for i ,j in list3:
    if i=='female'and j=='yes':
        femaelwhoSmoke+=1
    #else:
    elif i=='female'and j=='no':
            femaelwhoDontsmoke+=1
    elif i=='male'and j=='yes':
            maelwhoSmoke+=1
    elif i=='male'and j=='no':
            maelwhoDontsmoke+=1
   
        
        
print(maelwhoDontsmoke,"Male Who Don't Smoke")
print(maelwhoSmoke,"Male Who Smoke")
print(femaelwhoSmoke,"Femal Who Smoke")
print(femaelwhoDontsmoke,"Female Who Dont Smoke")
```

```python
"""Do an analysis of â€˜sexâ€™ and â€˜smokerâ€™ and comment on whether females smoke more 
than men or men smoke more than females, in max 30 words
with direct method 
here variable femaelwhoSmoke femael who smoke 
variable maelwhoSmoke mael who smoke 
variable femaelwhoDontsmoke femael who dont smoke 
variable maelwhoDontsmoke mael who dont smoke 
"""

femaelwhoSmoke=excel_red[(excel_red.sex=='female') & (excel_red.smoker=="yes")].count()
maelwhoSmoke=excel_red[(excel_red.sex=='male') & (excel_red.smoker=="yes")].count()
femaelwhoDontsmoke=excel_red[(excel_red.sex=='female') & (excel_red.smoker=="no")].count()
maelwhoDontsmoke=excel_red[(excel_red.sex=='male') & (excel_red.smoker=="no")].count()

if femaelwhoSmoke.id>maelwhoSmoke.id:
    
    print("Female Smoke more than Male")
else:
    print("Male Smoke more than Female")
```

```python

lisst={"femaelYes":femaelwhoSmoke.id,"maelYes":maelwhoSmoke.id,"femaelNo":femaelwhoDontsmoke.id,"maelNo":maelwhoDontsmoke.id}
lisst
```
